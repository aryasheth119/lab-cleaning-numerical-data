```python
def split_numerical_categorical(data):
    numericals = data.select_dtypes(include=[np.number])
    categoricals = data.select_dtypes(exclude=[np.number])
    return numericals, categoricals
```

```python
df = pd.DataFrame(data)
```

```python
df = df.drop('customer', axis=1)
print(df)
```

```python
def get_correlation_matrix(df):
    numerical_variables = df.select_dtypes(include='number')
    correlation_matrix = numerical_variables.corr()
    return correlation_matrix
```

```python
def get_highest_correlation(correlation_matrix):
    max_corr = correlation_matrix.unstack().sort_values(ascending=False)
    highest_corr_pair = max_corr[max_corr < 1].index[0]
    return highest_corr_pair
```

```python
def separate_numerical_columns(dataframe, threshold=36):
    continuous_cols = []
    discrete_cols = []

    for col in dataframe.columns:
        unique_values = dataframe[col].nunique()
        if unique_values > threshold:
            continuous_cols.append(col)
        else:
            discrete_cols.append(col)

    continuous_df = dataframe[continuous_cols]
    discrete_df = dataframe[discrete_cols]

    return continuous_df, discrete_df
```

```python
correlation_matrix = get_correlation_matrix(df)
highest_corr_pair = get_highest_correlation(correlation_matrix)
continuous_df, discrete_df = separate_numerical_columns(df)
```

```python
def create_barplots(dataframe):
    discrete_df = dataframe.select_dtypes(include='object')  

    fig, axes = plt.subplots(len(discrete_df.columns), 1, figsize=(16, 16))
    fig.subplots_adjust(hspace=0.5)  

    for i, column in enumerate(discrete_df.columns):
        sns.countplot(x=column, data=dataframe, ax=axes[i])
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('Count')

    plt.show()

create_barplots(discrete_df)
```

```python
def create_histograms(dataframe):
    continuous_df = dataframe.select_dtypes(include=['float64', 'int64'])  

    fig, axes = plt.subplots(len(continuous_df.columns), 1, figsize=(16, 16))
    fig.subplots_adjust(hspace=0.5)  

    for i, column in enumerate(continuous_df.columns):
        sns.histplot(dataframe[column], ax=axes[i])
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('Count')

    plt.show()

create_histograms(continuous_df)
```

```python
def create_boxplots(dataframe):
    continuous_df = dataframe.select_dtypes(include=['float64', 'int64'])  
    fig, axes = plt.subplots(len(continuous_df.columns), 1, figsize=(16, 16))
    fig.subplots_adjust(hspace=0.5) 

    for i, column in enumerate(continuous_df.columns):
        sns.boxplot(x=dataframe[column], ax=axes[i])
        axes[i].set_xlabel(column)

    plt.show()


create_boxplots(continuous_df)
```
